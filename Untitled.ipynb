```python
df = df.drop(columns = ['Unnamed: 0','LINE','FLO_LEAK_PHILLIPS','FLO_LEAK_RES_MED','FLO_LEAK_OTHER','RES_MED_PRES','PHILIPS_PRES','RECORDED_TIME','CONTACT_DATE'])
```

```python
df['ones']=1
```

```python
df_30 = df[df['DAYS_ON']<=30]
df_60 = df[(df['DAYS_ON']>30) & (df['DAYS_ON']<=60)]
df_90 = df[(df['DAYS_ON']>60) & (df['DAYS_ON']<=90)]
df_120 = df[(df['DAYS_ON']>90) & (df['DAYS_ON']<=120)]
df_150 = df[(df['DAYS_ON']>120) & (df['DAYS_ON']<=150)]
df_180 = df[(df['DAYS_ON']>150) & (df['DAYS_ON']<=180)]
```

```python
df_30_agg = df_30.groupby('MRN').agg({'FLO_AHI_RAW_VALUE': ['mean', 'std'],'FLO_FACE_TIME_VALUE': ['mean', 'std'],'Leak_Pressure': ['mean', 'std'],'Pressure': ['mean', 'std']})
df_60_agg = df_60.groupby('MRN').agg({'FLO_AHI_RAW_VALUE': ['mean', 'std'],'FLO_FACE_TIME_VALUE': ['mean', 'std'],'Leak_Pressure': ['mean', 'std'],'Pressure': ['mean', 'std']})
df_90_agg = df_90.groupby('MRN').agg({'FLO_AHI_RAW_VALUE': ['mean', 'std'],'FLO_FACE_TIME_VALUE': ['mean', 'std'],'Leak_Pressure': ['mean', 'std'],'Pressure': ['mean', 'std']})
df_120_agg = df_120.groupby('MRN').agg({'FLO_AHI_RAW_VALUE': ['mean', 'std'],'FLO_FACE_TIME_VALUE': ['mean', 'std'],'Leak_Pressure': ['mean', 'std'],'Pressure': ['mean', 'std']})
df_150_agg = df_150.groupby('MRN').agg({'FLO_AHI_RAW_VALUE': ['mean', 'std'],'FLO_FACE_TIME_VALUE': ['mean', 'std'],'Leak_Pressure': ['mean', 'std'],'Pressure': ['mean', 'std']})
df_180_agg = df_180.groupby('MRN').agg({'FLO_AHI_RAW_VALUE': ['mean', 'std'],'FLO_FACE_TIME_VALUE': ['mean', 'std'],'Leak_Pressure': ['mean', 'std'],'Pressure': ['mean', 'std']})
```

```python
df_30_agg['slope_AHI'] =0
df_30_agg['slope_facetime'] = 0
df_30_agg['slope_leak'] = 0
df_30_agg['slope_pres'] = 0
df_60_agg['slope_AHI'] = 0
df_60_agg['slope_facetime'] = 0
df_60_agg['slope_leak'] = 0
df_60_agg['slope_pres'] = 0
df_90_agg['slope_AHI'] = 0
df_90_agg['slope_facetime'] = 0
df_90_agg['slope_leak'] = 0
df_90_agg['slope_pres'] = 0
df_120_agg['slope_AHI'] = 0
df_120_agg['slope_facetime'] = 0
df_120_agg['slope_leak'] = 0
df_120_agg['slope_pres'] = 0
df_150_agg['slope_AHI'] = 0
df_150_agg['slope_facetime'] = 0
df_150_agg['slope_leak'] = 0
df_150_agg['slope_pres'] = 0
df_180_agg['slope_AHI'] = 0
df_180_agg['slope_facetime'] = 0
df_180_agg['slope_leak'] = 0
df_180_agg['slope_pres'] = 0
```

```python
names = df['MRN'].unique()
cat = ['FLO_AHI_RAW_VALUE','FLO_FACE_TIME_VALUE','Leak_Pressure','Pressure']
slope = ['slope_AHI','slope_facetime','slope_leak','slope_pres']
```

```python
for j in range(4):
    for i in names:
        df1 = df_30[df_30['MRN']==i].reset_index()
        df2 = df_60[df_60['MRN']==i].reset_index()
        df3 = df_90[df_90['MRN']==i].reset_index()
        df4 = df_120[df_120['MRN']==i].reset_index()
        df5 = df_150[df_150['MRN']==i].reset_index()
        df6 = df_180[df_180['MRN']==i].reset_index()
        A1 = df1[['DAYS_ON','ones']]
        A2 = df2[['DAYS_ON','ones']]
        A3 = df3[['DAYS_ON','ones']]
        A4 = df4[['DAYS_ON','ones']]
        A5 = df5[['DAYS_ON','ones']]
        A6 = df6[['DAYS_ON','ones']]
        y1 = df1[cat[j]]
        y2 = df2[cat[j]]
        y3 = df3[cat[j]]
        y4 = df4[cat[j]]
        y5 = df5[cat[j]]
        y6 = df6[cat[j]]
        m1,c = np.linalg.lstsq(A1,y1)[0]
        m2,c = np.linalg.lstsq(A2,y2)[0]
        m3,c = np.linalg.lstsq(A3,y3)[0]
        m4,c = np.linalg.lstsq(A4,y4)[0]
        m5,c = np.linalg.lstsq(A5,y5)[0]
        m6,c = np.linalg.lstsq(A6,y6)[0]
        df_30_agg.loc[i,slope[j]] = m1
        df_60_agg.loc[i,slope[j]] = m2
        df_90_agg.loc[i,slope[j]] = m3
        df_120_agg.loc[i,slope[j]] = m4
        df_150_agg.loc[i,slope[j]] = m5
        df_180_agg.loc[i,slope[j]] = m6
```
