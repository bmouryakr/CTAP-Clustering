```python
def scaled(df,df1):
    X = df1.values
    scaler = StandardScaler()
    scaler.fit(X)
    X_scaled = scaler.transform(df.values)
    df_scaled = pd.DataFrame(data=X_scaled,columns = df.columns, index = df.index)
    return df_scaled
```

```python
def plotcurve_gm(df):
    n_components = np.arange(1, 21)
    models = [mix(n, random_state=0).fit(df) for n in n_components]
    plt.plot(n_components, [m.bic(df) for m in models], label='BIC',linewidth=4)
    plt.plot(n_components, [m.aic(df) for m in models], label='AIC',linewidth=4)
    plt.title('GMM - AIC,BIC vs no. of clusters',fontsize=16)
    plt.legend(loc='best',fontsize=14)
    plt.xlabel('number of clusters',fontsize=14)
    plt.xticks(list(range(2,21,2)),fontsize=14)
    plt.yticks(fontsize=14)
    plt.savefig('/Users/mkr4014/Desktop/Academics/CTAP/AIC-BIC.png');
```

```python
def plotcurve_KM(df):
    n_components = np.arange(1, 21)
    models = [KM(n, random_state=0).fit(df) for n in n_components]
    plt.plot(n_components, [-m.score(df) for m in models], label='SS',linewidth=4)
    plt.title('K-Means - SS vs no. of clusters',fontsize=16)
    plt.ylabel('Sum of Squares',fontsize=14)
    plt.xlabel('number of clusters',fontsize=14)
    plt.xticks(list(range(2,21,2)),fontsize=14)
    plt.yticks(fontsize=14)
    plt.savefig('/Users/mkr4014/Desktop/Academics/CTAP/SS.png');
```

```python
def calc_probabilities(df_list):
    n = len(df_list)
    clusters = len(df_list[0].iloc[:,-1].unique())
    prob = np.zeros([clusters,clusters])
    denom = np.zeros(clusters)
    for i in range(n-1):
        df1 = pd.DataFrame(df_list[i].iloc[:,-1])
        df2 = pd.DataFrame(df_list[i+1].iloc[:,-1])
        for j in range(clusters):
            df3 = df1[df1.iloc[:,0]==j]
            denom[j] += len(df3)
            df4 = df3.join(df2,how='inner')
            for k in range(clusters):
                l = len(df4[df4.iloc[:,-1]==k])
                prob[j,k] += l
    for i in range(clusters):
        prob[i]=prob[i]/denom[i]
    return prob
```

```python
def find_states(prob):
    states = np.zeros(prob.shape[0])
    I = np.identity(prob.shape[0])
    prob = prob.T-I
    prob[0]=np.ones(prob.shape[0])
    states[0] = 1
    state_prob = np.linalg.solve(prob,states)
    return state_prob
```

```python
def calc_probabilities_ind(df_list):
    n = len(df_list)-1
    clusters = len(df_list[0].iloc[:,-1].unique())
    prob = np.zeros([n,clusters,clusters])
    for i in range(n):
        df1 = pd.DataFrame(df_list[i].iloc[:,-1])
        df2 = pd.DataFrame(df_list[i+1].iloc[:,-1])
        for j in range(clusters):
            df3 = df1[df1.iloc[:,0]==j]
            denom = len(df3)
            df4 = df3.join(df2,how='inner')
            for k in range(clusters):
                l = len(df4[df4.iloc[:,-1]==k])
                prob[i,j,k] = l/denom
    return prob
```
